```python
#Changing the column names
data = data.rename(columns={'Age (in years)':'Age','Experience (in years)':'Experience', 'Income (in K/month)': 'Income',
                                  'Family members':'Family_members', 'Personal Loan' : 'Personal_Loan',
                           'Securities Account' : 'Securities_Account','CD Account' : 'CD_Account'})
data.columns
```

```python
data.head()
```

```python
df = data.dropna() # Dropping null values in Family members
```

```python
df.isnull().sum()
```

```python
df = df.drop(['ID','ZIP Code','Experience','Family_members','Online','CreditCard','Age','Mortgage'],axis=1)  
#Dropping features which do not have major impact on Target column
# Removing Experience and Age column as it is highly correlated and forms negative correlation with Target column
```

```python
df.head()
```

## Model Building

```python
# Importing required libraries
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
#Dropping the target column from X
X = df.drop(columns = 'Personal_Loan')
y = df['Personal_Loan']
```

```python
#Splitting the dataset into training and test set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.30, random_state= 42)
```

```python
X_train.shape ,X_test.shape ,y_train.shape, y_test.shape
```

```python
#Model building
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train,y_train)
```

```python
tree_pred = tree_model.predict(X_test)
```

```python
tree_model.score(X_test,y_test)     
```

```python
print(classification_report(y_test, tree_pred))
```

```python
print(confusion_matrix(y_test, tree_pred))
```

## Hyperparameter Tuning

```python
#I am tuning three hyperparameters right now, amd I am passing the different values for both parameters
grid_param = {
    'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,32,1),
    'min_samples_leaf' : range(1,10,1),
    'min_samples_split': range(2,10,1),
    'splitter' : ['best', 'random']   
}
```

```python
grid_search_tree = GridSearchCV(estimator=tree_model,
                     param_grid=grid_param,
                     cv=5,
                    n_jobs =-1)
```

```python
grid_search_tree.fit(X_train,y_train)
```

```python
best_parameters = grid_search_tree.best_params_
print(best_parameters)
```

```python
Tree_accuracy = grid_search_tree.best_score_
```

```python
print('Decision Tree Model Accuracy:', Tree_accuracy * 100, '%')
```

After applying Hyperparameter turning ,the accuracy got improved from 96% to 97%
